# Chart

```python
import matplotlib.pyplot as plt
```

```python
label_counts = df_ori['label'].value_counts()
label_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Labels (prev. Research)')
plt.show()
```

```python
# Define a function to format the labels
def custom_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 0 else ''

plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(label_counts, labels=label_counts.index, autopct=custom_autopct, startangle=140)

# Adjust label size
for text in texts:
    text.set_fontsize(16)  # Adjust as needed
for autotext in autotexts:
    autotext.set_fontsize(16)  # Adjust as needed

plt.title('Distribution of Labels (prev. Research)', fontsize=18)
plt.show()
```

```python
# Plot the treemap
sizes = label_counts.values
labels = [f'{label}\n{size}' for label, size in zip(label_counts.index, sizes)]

plt.figure(figsize=(8, 8))
squarify.plot(sizes=sizes, label=labels, alpha=.8)
plt.title('Distribution of Labels')
plt.axis('off')
plt.show()
```

# Wordcloud

```python
from wordcloud import WordCloud
```

```python
# Combine all text data into a single string
text_combined = ' '.join(df_ori['text'])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text_combined)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud from Text Data (prev. Research)', fontsize=16)
plt.show()
```
