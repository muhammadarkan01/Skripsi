## Satuin File di Folder dataset

```python
import glob
 
# list all csv files only
# csv_files = glob.glob('*.{}'.format('csv'))
# csv_files

# Specify the folder or directory path
folder_path = "../1 Scraping/Dataset"

# Find all text files in the specified folder and its subdirectories
csv_files = glob.glob(folder_path + "/*.csv")
csv_files
```

```python
import pandas as pd
 
df_csv_append = pd.DataFrame()

#Append files
for file in csv_files:
    df = pd.read_csv(file)
    # Remove the "Unnamed: 0" column if present
    if "Unnamed: 0" in df.columns:
        df.drop("Unnamed: 0", axis=1, inplace=True)
    df_csv_append = df_csv_append.append(df, ignore_index=True)

df_csv_append
```

```python
df_csv_append.to_csv('scraped_tweets.csv')
```
